```python
import numpy as np
import pandas as pd
import plotly.express as px


df = pd.read_csv(r'..\data\filter_ja_zip_data.csv', nrows=100)

#view the first 10 rows
#df.head(10)
#if columns are many and unable to see in JP Notebook view
df.head(10).T
```

```python
import pandas as pd
import re

# Assuming you have an existing DataFrame named 'df' with your dataset
# Create a copy of the DataFrame to avoid overwriting the original data
df_copy = df.copy()

# Function to extract latitude and longitude using regex and handle null values
def extract_coordinates(cell):
    if cell is not None and isinstance(cell, str):
        match = re.findall(r'POINT \((.*?) (.*?)\)', cell)
        if match:
            return float(match[0][1]), float(match[0][0])
    return None, None

# Apply the extraction function to create "Latitude" and "Longitude" columns in the copy
df_copy['Latitude'], df_copy['Longitude'] = zip(*df_copy['Location Point'].map(extract_coordinates))


# Display the resulting DataFrame (df_copy)
print(df_copy)

df_copy.head(10).T
```

```python
#deleteing some cells to clean up

df_copy.drop(columns=['Section ID', 'Row', 'Site Number'], inplace=True)

df_copy.head(10).T
```

```python
#deleteing some cells to clean up

df_copy.drop(columns=['Relationship to Veteran', 'First Name of Related Veteran',\
    'Middle Name of Related Veteran', 'Last Name of Related Veteran', 'Suffix of Related Veteran'], inplace=True)

df_copy.head(10).T
```

```python
column_data_types = df_copy.dtypes

# Display the data types of all columns
print(column_data_types)
```

```python
# Convert date columns to datetime
df_copy['Birth Date of Deceased'] = pd.to_datetime(df_copy['Birth Date of Deceased'])
df_copy['Death Date of Deceased'] = pd.to_datetime(df_copy['Death Date of Deceased'])
df_copy['last_update_date'] = pd.to_datetime(df_copy['last_update_date'])


column_data_types = df_copy.dtypes

# Display the data types of all columns
print(column_data_types)

df_copy.head(10).T
```

```python
# Extract the column you want to move
last_update_date_column = df_copy['last_update_date']

# Drop the column from its current position
df_copy.drop(columns=['last_update_date'], inplace=True)

# Reinsert the column at the end of the DataFrame
df_copy['last_update_date'] = last_update_date_column

# Display the resulting DataFrame with "last_update_date" as the last column
df_copy.head(10).T
```

```python
#removing zeros for time in the date columns
df_copy['Birth Date of Deceased'] = df_copy['Birth Date of Deceased'].dt.strftime('%Y-%m-%d')
df_copy['Death Date of Deceased'] = df_copy['Death Date of Deceased'].dt.strftime('%Y-%m-%d')
df_copy['last_update_date'] = df_copy['last_update_date'].dt.strftime('%Y-%m-%d')

df_copy.head(10).T
```
