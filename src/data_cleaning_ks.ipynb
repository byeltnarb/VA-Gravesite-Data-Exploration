DATA PROJECT FOR DECEASED VETERANS GRAVESITE LOCATION
A project for the local legion post 44 of Jackson County, KS.

```python
import numpy as np
import pandas as pd

#exploring size of data for KS specifically ZipCodes 66416, 66418, 66419, 66436, 66440, 66509, 66516, 66540, 66552
#recieved from https://www.zip-codes.com/county/ks-jackson.asp and last updated August 1, 2023

df_entire = pd.read_csv(r'..\data\National_Cemetery_Administration_Gravesite_Locator_VA_dataset.csv', dtype={'Cemetery ZIP': str}, low_memory = False)

specific_zips_entire = ['66416', '66418', '66419', '66436', '66440', '66509', '66516', '66540', '66552']

# Filter and count the specified zip codes
filtered_counts_entire = df_entire['Cemetery ZIP'].value_counts()[specific_zips_entire]
total_count_entire = filtered_counts_entire.sum()

print(filtered_counts_entire)
print("Total count for specified zip codes: ", total_count_entire)
```

```python
print(df_entire['Cemetery ZIP'].dtype)

print(df_entire[['Cemetery ZIP', 'Death Date of Deceased', 'last_update_date']].dtypes)

```

```python
# applying datatime to columns with date

date_columns = ['Birth Date of Deceased', 'Death Date of Deceased', 'last_update_date']

for col in date_columns:
    df_entire[col] = pd.to_datetime(df_entire[col], errors='coerce')
```

```python
print(df_entire['Cemetery ZIP'].dtype)

print(df_entire[['Cemetery ZIP', 'Birth Date of Deceased', 'Death Date of Deceased', 'last_update_date']].dtypes)
```

```python
filtered_df = df_entire[df_entire['Cemetery ZIP'].isin(specific_zips_entire)]
filtered_df.head(10).T


```

```python
specific_ks = ['KS']
ks_counts = df_entire['Cemetery State'].value_counts()[specific_ks]
print(ks_counts)
```

```python
filtered_ks = df_entire[df_entire['Cemetery State'].isin(specific_ks)]
filtered_ks.head(10).T
```

```python
#KS to csv in the data folder
filtered_ks.to_csv('../data/filter_ks_data.csv', index=False)
```
