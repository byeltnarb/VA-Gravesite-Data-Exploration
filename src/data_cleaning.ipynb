CLEANING the ZIP CODE column

```python
#NaN values to 99999

df_entire['Cemetery ZIP'].fillna('99999', inplace=True)
```

```python
# convert all values to strings

df_entire['Cemetery ZIP'] = df_entire['Cemetery ZIP'].astype(str)
```

```python
#remove ".0" from the end

df_entire['Cemetery ZIP'] = df_entire['Cemetery ZIP'].astype(str).str.replace('\.0$', '', regex=True)

df_entire['Cemetery ZIP'].sample(20)  # Displays a random sample of 50 ZIP codes
```

```python
#append 'X' character to the end of short US Zip Codes with standard 5-digit length (possibly Australia)

def pad_zip(zip_code):
    if len(zip_code) < 5 and zip_code.isdigit():
        return zip_code + 'X' * (5 - len(zip_code))
    return zip_code

df_entire['Cemetery ZIP'] = df_entire['Cemetery ZIP'].apply(pad_zip)
```

```python
#reverting the 'X' due to posisble Autstralia Zip Codes

def revert_padded_zip(zip_code):
    # If zip_code ends with 'X', revert the padding.
    while zip_code.endswith('X'):
        zip_code = zip_code[:-1]
    return zip_code

df_entire['Cemetery ZIP'] = df_entire['Cemetery ZIP'].apply(revert_padded_zip)
```
