```python
#exploring size of data for KS specifically ZipCodes 66416, 66418, 66419, 66436, 66440, 66509, 66516, 66540, 66552
#recieved from https://www.zip-codes.com/county/ks-jackson.asp and last updated August 1, 2023

df_entire = pd.read_csv(r'..\data\National_Cemetery_Administration_Gravesite_Locator_VA_dataset.csv')

#converting the column to string - based on the inspecting 
df_entire['Cemetery ZIP'] = df_entire['Cemetery ZIP'].astype(str)

specific_zips_entire = [66416,66418,66419,66436,66440,66509,66516,66540,66552]
# Convert specific_zips_entire to strings
specific_zips_str = [str(zip_code) for zip_code in specific_zips_entire]

zip_counts_entire = df_entire['Cemetery ZIP'].value_counts()
filtered_counts_entire = zip_counts_entire[zip_counts_entire.index.isin(specific_zips_str)]
total_count_entire = filtered_counts_entire.sum()

print(filtered_counts_entire)
print("Total count for specified zip codes: ",total_count_entire)


```

```python
# Filter the DataFrame to only include rows with those ZIP codes
filtered_df = df_entire[df_entire['Cemetery ZIP'].isin(specific_zips_str)]

# Display the filtered DataFrame
print(filtered_df)
filtered_df
```

```python
#filtered ZIP CODE to CSV
filtered_df.to_csv('filtered_data.csv', index=False)
```
